## Chapter 1 : Reservoir Computing for chaotic timeseries forecasting <span id="chapitre1"/>

**Mackey-Glass timeseries**

Mackey-Glass equation are a set of delayed differential equations
describing the temporal behaviour of different physiological signal,
for example, the relative quantity of mature blood cells over time.

The equations are defined as:

$$
\frac{dP(t)}{dt} = \frac{a P(t - \tau)}{1 + P(t - \tau)^n} - bP(t)
$$

where $a = 0.2$, $b = 0.1$, $n = 10$, and the time delay $\tau = 17$.
$\tau$ controls the chaotic behaviour of the equations (the higher it is,
the more chaotic the timeserie becomes.
$\tau=17$ already gives good chaotic results.)

### Setup

We setup the code by fixing the random seed of the notebook, to ensure reproducibility, using the :py:func:`~.set_seed` function.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import reservoirpy as rpy

rpy.set_seed(42)
```

### Loading data

The :py:mod:`~.datasets` module contains several usefull timeseries generator such as the :py:func:`~.mackey_glass` function. We use this function to generate 3000 timesteps of the Mackey-Glass timeseries, with a $\tau$ parameter equals to 17, to obtain interesting chaotic behavior.

```python
from reservoirpy.datasets import mackey_glass
from reservoirpy.observables import nrmse, rsquare

timesteps = 3000
tau = 17
X = mackey_glass(timesteps, tau=tau)

# rescale between -1 and 1
X = 2 * (X - X.min()) / (X.max() - X.min()) - 1
```

```python
def plot_mackey_glass(X, sample, tau):

    fig = plt.figure(figsize=(13, 5))
    N = sample

    ax = plt.subplot((121))
    t = np.linspace(0, N, N)
    for i in range(N-1):
        ax.plot(t[i:i+2], X[i:i+2], color=plt.cm.magma(255*i//N), lw=1.0)

    plt.title(f"Timeseries - {N} timesteps")
    plt.xlabel("$t$")
    plt.ylabel("$P(t)$")

    ax2 = plt.subplot((122))
    ax2.margins(0.05)
    for i in range(N-1):
        ax2.plot(X[i:i+2], X[i+tau:i+tau+2], color=plt.cm.magma(255*i//N), lw=1.0)

    plt.title(f"Phase diagram: $P(t) = f(P(t-\\tau))$")
    plt.xlabel("$P(t-\\tau)$")
    plt.ylabel("$P(t)$")

    plt.tight_layout()
    plt.show()
```

```python
plot_mackey_glass(X, 500, tau)
```

The Mackey-Glass timeseries is:

- Not completely unpredictable... (not random)
- ...but not easily predictable (not periodic)
- Similar to ECG rythms, stocks, weather...

This is a typical use case for Reservoir Computing techniques such as **Echo State Networks (ESN)**.

##### Data preprocessing

We split the timeseries in two: a training set and a testing set. Training dataset is used to train the network, while testing dataset is used to evaluate its performance on data it has never *seen* before.

```python
def plot_train_test(X_train, y_train, X_test, y_test):
    sample = 500
    test_len = X_test.shape[0]
    fig = plt.figure(figsize=(15, 5))
    plt.plot(np.arange(0, 500), X_train[-sample:], label="Training data")
    plt.plot(np.arange(0, 500), y_train[-sample:], label="Training objective")
    plt.plot(np.arange(500, 500+test_len), X_test, label="Testing data")
    plt.plot(np.arange(500, 500+test_len), y_test, label="Testing objective")
    plt.legend()
    plt.show()
```

##### Build your first Echo State Network

While Reservoir Computing techniques rely on training only a very few connections in the network, most architectures need to be properly tuned using different hyperparameters to obtain interesting results.

In this tutorial, we will tune:

- `units`: the number of neurons in the reservoir. The more there is, the best results you obtain, the more overfitting you get.
- `leak_rate`: the neurons time constant. The lower it is, the more inertia you get in the reservoir activity.
- `spectral_radius`: the recurrent matrix spectral radius. This parameter controls the chaoticity of reservoir activity, and can be quite complex to tune.
- `connectivity`: the proportion of available connections used in the reservoir. Should be quite low.
- `input_connectivity`: the proportion of available connecitons used between the reservoir and the inputs. Should be quite low also.
- `regularization`: the L2 regularization coefficient used to increase generalization of the training using linear regression.
- `seed`: a random generator seed. As most connections in an ESN are randomly drawn, you should pay extra attention to the reproducibility of your results, by averaging performance over several instances of networks with different random initializations.

```python
units = 100
leak_rate = 0.3
spectral_radius = 1.25
input_scaling = 1.0
connectivity = 0.1
input_connectivity = 0.2
regularization = 1e-8
seed = 1234
```

To create an ESN, we then need two ReservoirPy objects instances, called *Nodes*: a :py:class:`~.Reservoir` node
and a :py:class:`~.Ridge` node.

```python
# Some function to create a new ESN when needed
def reset_esn():
    from reservoirpy.nodes import Reservoir, Ridge

    reservoir = Reservoir(units, input_scaling=input_scaling, sr=spectral_radius,
                          lr=leak_rate, rc_connectivity=connectivity,
                          input_connectivity=input_connectivity, seed=seed)
    readout   = Ridge(1, ridge=regularization)

    return reservoir >> readout
```

#### Entranement de l'ESN

L'apprentissage est *offline* ("hors-ligne") : il n'a lieu qu'une seule fois, sur l'ensemble des donnes d'entranement.

```python
esn = esn.fit(X_train1, y_train1)
```

```python
def plot_readout(readout):
    Wout = readout.Wout
    bias = readout.bias
    Wout = np.r_[bias, Wout]

    fig = plt.figure(figsize=(15, 5))

    ax = fig.add_subplot(111)
    ax.grid(axis="y")
    ax.set_ylabel("Coefs. de $W_{out}$")
    ax.set_xlabel("Neurones du reservoir")
    ax.bar(np.arange(Wout.size), Wout.ravel()[::-1])

    plt.show()
```

```python
plot_readout(readout)
```

#### Test de l'ESN

```python
def plot_results(y_pred, y_test, sample=500):

    fig = plt.figure(figsize=(15, 7))
    plt.subplot(211)
    plt.plot(np.arange(sample), y_pred[:sample], lw=3, label="ESN prediction")
    plt.plot(np.arange(sample), y_test[:sample], linestyle="--", lw=2, label="True value")
    plt.plot(np.abs(y_test[:sample] - y_pred[:sample]), label="Absolute deviation")

    plt.legend()
    plt.show()
```

```python
y_pred1 = esn.run(X_test1)
```

```python
plot_results(y_pred1, y_test1)
```

Coefficient de dtermination $R^2$ et erreur quadratique normalise :

```python
rsquare(y_test1, y_pred1), nrmse(y_test1, y_pred1)
```

```python
units = 500
leak_rate = 0.3         # - leaking rate
spectral_radius = 0.99  # - rayon spectral
input_scaling = 1.0     # - facteur de mise  l'chelle des entres (input scaling)
connectivity = 0.1      # - densit des connexions du reservoir vers lui mme
input_connectivity = 0.2  # et des entres vers le reservoir
regularization = 1e-4   # - coefficient de rgularisation (L2)
seed = 1234             # reproductibilit
```

```python
def plot_generation(X_gen, X_t, nb_generations, warming_out=None, warming_inputs=None, seed_timesteps=0):

    plt.figure(figsize=(15, 5))
    if warming_out is not None:
        plt.plot(np.vstack([warming_out, X_gen]), label="Srie gnre")
    else:
        plt.plot(X_gen, label="Srie gnre")

    plt.plot(np.arange(nb_generations)+seed_timesteps, X_t, linestyle="--", label="Srie relle")

    if warming_inputs is not None:
        plt.plot(np.arange(seed_timesteps), warming_inputs, linestyle="--", label="Donnes d'chauffement")

    plt.plot(np.arange(nb_generations)+seed_timesteps, np.abs(X_t - X_gen),
             label="Erreur absolue")

    if seed_timesteps > 0:
        plt.fill_between([0, seed_timesteps], *plt.ylim(), facecolor='lightgray', alpha=0.5, label="Echauffement")

    plt.plot([], [], ' ', label=f"$R^2 = {round(rsquare(X_t, X_gen), 4)}$")
    plt.plot([], [], ' ', label=f"$NRMSE = {round(nrmse(X_t, X_gen), 4)}$")
    plt.legend(
    )
    plt.show()
```

####Entranement  la prvision sur un horizon court

```python
esn = reset_esn()

x, y = to_forecasting(X, forecast=1)
X_train3, y_train3 = x[:2000], y[:2000]
X_test3, y_test3 = x[2000:], y[2000:]

esn = esn.fit(X_train3, y_train3)
```

####Gnration

- 100 pas de temps de la srie de test utiliss comme "chauffement";
- 300 pas de temps gnrs " partir de rien";

```python
seed_timesteps = 100

warming_inputs = X_test3[:seed_timesteps]

warming_out = esn.run(warming_inputs, reset=True)  # chauffement
```

```python
nb_generations = 400

X_gen = np.zeros((nb_generations, 1))
y = warming_out[-1]
for t in range(nb_generations):  # gnration
    y = esn(y)
    X_gen[t, :] = y
```

```python
X_t = X_test3[seed_timesteps: nb_generations+seed_timesteps]
plot_generation(X_gen, X_t, nb_generations, warming_out=warming_out,
                warming_inputs=warming_inputs, seed_timesteps=seed_timesteps)
```

## Aller plus loin : Comprendre les hyperparamtres et leurs effets <span id="bonus"/>

```python
units = 100             # - nombre de neurones dans le reservoir
leak_rate = 0.3         # - leaking rate
spectral_radius = 1.25  # - rayon spectral
input_scaling = 1.0     # - facteur de mise  l'chelle des entres
connectivity = 0.1      #- densit des connexions du reservoir vers lui mme
input_connectivity = 0.2  # et des entres vers le reservoir
regularization = 1e-8   # - coefficient de rgularisation (L2)
seed = 1234             # reproductibilit
```

### 1. Le rayon spectral

Le rayon spectral est **la valeur propre maximale de la matrice des poids du rservoir ($W$)**.

```python
states = []
radii = [0.1, 1.25, 10.0]
for sr in radii:
    reservoir = Reservoir(units, sr=sr, input_scaling=0.1, lr=leak_rate, rc_connectivity=connectivity,
                         input_connectivity=input_connectivity)

    s = reservoir.run(X_test1[:500])
    states.append(s)
```

```python
units_nb = 20

plt.figure(figsize=(15, 8))
for i, s in enumerate(states):
    plt.subplot(len(radii)*100+10+i+1)
    plt.plot(s[:, :units_nb], alpha=0.6)
    plt.ylabel(f"$sr={radii[i]}$")
plt.xlabel(f"Activations ({units_nb} neurons)")
plt.show()
```

- $-$ rayon spectral $\rightarrow$ dynamiques **stables**

- $+$ rayon spectral $\rightarrow$ dynamiques **chaotiques**

Rayon spectral et *Echo State Property* : rayon spectral $\rightarrow$ 1 (assure que les tats internes ne sont pas affects par l'initialisation).

### 2. Le facteur de mise  l'chelle des entres (*input scaling*)

Il s'agit d'un **coefficient appliqu  $W_{in}$**, venant changer l'chelle des donnes en entre.

```python
states = []
scalings = [0.01, 0.1, 1.]
for iss in scalings:
    reservoir = Reservoir(units, sr=spectral_radius, input_scaling=iss, lr=leak_rate,
                          rc_connectivity=connectivity, input_connectivity=input_connectivity)

    s = reservoir.run(X_test1[:500])
    states.append(s)
```

```python
def correlation(states, inputs):
    return np.mean([np.correlate(states[:, i].flatten(), inputs.flatten()) for i in range(states.shape[1])])
```

```python
units_nb = 20

plt.figure(figsize=(15, 8))
for i, s in enumerate(states):
    plt.subplot(len(scalings)*100+10+i+1)
    plt.plot(s[:, :units_nb], alpha=0.6)
    plt.ylabel(f"$iss={scalings[i]}$")
plt.xlabel(f"Activations ({units_nb} neurons)")
plt.show()
```

Correlation moyenne des activits des neurones du reservoir avec les entres :

```python
for i, s in enumerate(states):
    corr = correlation(states[i], X_test1[:500])
    print(f"ISS : {scalings[i]}, correlation moyenne : {corr}")
```

- $+$ input scaling $\rightarrow$ activits **corrles aux donnes**
- $-$ input scaling $\rightarrow$ activits **libres**

L'*input scaling* peut aussi tre utilis pour ajuster l'influence de chaque donne en entre.

### 3. Le taux de fuite de la mmoire (*leaking rate*)


$$
x(t+1) =  \underbrace{\color{red}{(1 - \alpha)} x(t)}_{\text{tat actuel}} + \underbrace{\color{red}\alpha f(u(t+1), x(t))}_{\text{donnes suivantes}}
$$

avec $\alpha \in [0, 1]$ et:

$$ f(u, x) = \tanh(W_{in} \cdotp u + W \cdotp x) $$

```python
states = []
rates = [0.02, 0.2, 0.9]
for lr in rates:
    reservoir = Reservoir(units, sr=spectral_radius, input_scaling=input_scaling, lr=lr,
                          rc_connectivity=connectivity, input_connectivity=input_connectivity)

    s = reservoir.run(X_test1[:500])
    states.append(s)
```

```python
units_nb = 20

plt.figure(figsize=(15, 8))
for i, s in enumerate(states):
    plt.subplot(len(rates)*100+10+i+1)
    plt.plot(s[:, :units_nb] + 2*i)
    plt.ylabel(f"$lr={rates[i]}$")
plt.xlabel(f"States ({units_nb} neurons)")
plt.show()
```

- $+$ leaking rate $\rightarrow$ **faible inertie**, faible mmorisation des tats prcdents
- $-$ leaking rate $\rightarrow$ **forte inertie**, grande mmorisation des tats prcdents

Le *leaking rate* contrle la "mmoire" de l'ESN. Il peut tre vu comme l'inverse de sa constante de temps
